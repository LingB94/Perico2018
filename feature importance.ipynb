```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import sklearn
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('2018.csv')
fea = data.iloc[:, 1:-1]
y = data['result']
features = fea.columns
X = data[features]
```

Feature Importance Ranking

```python
from sklearn.ensemble import ExtraTreesClassifier
first = set([12,11,15,14,5,4,6,9,10,1])
model = ExtraTreesClassifier()
model.fit(X, y)
ETC_feature_importances = pd.DataFrame(model.feature_importances_, index = X.columns, columns=['ETC']).sort_values('ETC', ascending=False)
print(ETC_feature_importances)
```

```python
attr_sort = ['The location of the disease','Gender','The height type of the posterior cushion of the molar','Whether the root of the third molar is deformed abnormally','Visit month','The positional relationship between the third molar and the mandibular neural tube','Age',' The slope of the posterior molar pad','The ratio of the crown and root of the third mandibular molar','The inclination of the third maxillary molar','The type of mandibular third molar','Mandibular third molar eruption direction','Location relationship between the third molar and adjacent tooth','Whether the lesions of the third molar cause damage to adjacent teeth','Treatment strategy']
```

```python
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
ETC_importance = model.feature_importances_
indices = np.argsort(ETC_importance)
range_ = range(len(ETC_importance))
plt.figure(figsize = (10, 7))
plt.title("feature importance")
plt.barh(range_, ETC_importance[indices],color='r',alpha=0.8,align='center')
for x, y in enumerate(np.around(ETC_importance[indices], decimals = 7)):
    plt.text(y, x, str(y), color='black', fontweight='light')
plt.yticks(range_,attr_sort)
plt.ylim([-1,len(ETC_importance)])
plt.xlim([0.0,0.5])
plt.show()
```
